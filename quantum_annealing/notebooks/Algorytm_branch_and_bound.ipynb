```python
import numpy as np
from collections import namedtuple

Node = namedtuple("Node", ["partial_solution", "deph"])

# funkcje pomocnicze


def branch_and_bound(J, h):
    n = len(h)

    stack = {-1: [[]]}

    for i in range(n):
        deph = stack[i - 1]
        temp = []
        for branch in deph:
            p = h[i]
            c = J[i, :]
            plus = branch + [1]
            minus = branch + [-1]
            
            if np.sum(np.abs(c)) <= np.abs(p):
                max_d = np.sum(np.abs(c)) + p
                min_d = - np.sum(np.abs(c)) + p
                if max_d <= 0:
                    temp.append(plus)
                elif min_d >= 0:
                    temp.append(minus)
                print("pruning")
            else:
                temp.append(plus)
                temp.append(minus)
            # stack[i + 1] = stack[i].append(minus)
        stack[i] = temp
        
        

        

        
         
```
