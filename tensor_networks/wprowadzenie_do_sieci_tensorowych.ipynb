# Wprowadzenie do sieci tensorowych

Jako tensor będziemy rozumieli indeksowany zbiór liczb.

![TN]()

Sieci tensorowe to matematyczne struktury służące do efektywnej reprezentacji i manipulacji wielowymiarowych danych, często stosowane w fizyce kwantowej oraz uczeniu maszynowym. Formalnie, sieć tensorowa definiowana jest jako zbiór tensorów połączonych indeksami, po których wykonywane jest sumowanie (kontrakcja).

Rozważmy układ $ N $ tensorów $\{T^{(1)}, T^{(2)}, \dots, T^{(N)}\}$, gdzie każdy tensor $T^{(k)}$ posiada zbiór indeksów. Sieć tensorowa powstaje przez kontrakcję wybranych indeksów tych tensorów:
$$
\mathcal{F}_{i_1, i_2, \dots, i_n} = \sum_{\alpha, \beta, \dots} T^{(1)}_{i_1, \alpha} T^{(2)}_{\alpha, i_2, \beta} \dots T^{(N)}_{\dots, i_n}.
$$

Podstawowymi przykładami takich struktur są:

* **MPS** (Matrix Product States) 
* **PEPS** (Projected Entangled Pair States)


Sieci tensorowe pozwalają na kompaktową reprezentację dużych przestrzeni Hilberta, umożliwiając efektywne obliczenia numeryczne i analityczne w wielu dziedzinach nauki.
