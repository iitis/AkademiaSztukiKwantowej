# Wyprowadzenie wzoru na różnicę energii

Przyjmijmy konfigurację spinów

$$
s=(s_1,\dots,s_N), \qquad s_i\in\{-1,+1\}.
$$

Używamy konwencji hamiltonianu

$$
E(s)= -\frac12 \sum_{i,j} J_{ij}s_i s_j - \sum_i h_i s_i.
$$

Chcemy policzyć różnicę energii między dwoma stanami, które różnią się tylko jednym spinem, powiedzmy spinem o numerze $k$.

## 1. Definicja dwóch stanów

Niech stan $s'$ powstaje ze stanu $s$ przez odwrócenie tylko spinu $k$:

$$
s'_i=
\begin{cases}
s_i, & i\neq k, \\
-s_k, & i=k.
\end{cases}
$$

Interesuje nas:

$$
\Delta E = E(s')-E(s).
$$

## 2. Rozpisanie energii obu stanów

Mamy

$$
E(s)= -\frac12 \sum_{i,j} J_{ij}s_i s_j - \sum_i h_i s_i,
$$

oraz

$$
E(s')= -\frac12 \sum_{i,j} J_{ij}s'_i s'_j - \sum_i h_i s'_i.
$$

Odejmujemy:

$$
\Delta E
=
-\frac12 \sum_{i,j} J_{ij}(s'_i s'_j-s_i s_j)
-\sum_i h_i(s'_i-s_i).
$$

Ponieważ zmienia się tylko spin $k$, wszystkie wyrazy niezawierające indeksu $k$ skracają się.

## 3. Oddziaływania spin-spin (J)

Zostają tylko składniki, w których występuje $k$, czyli te z $i=k$ albo $j=k$:

$$
\Delta E_{\mathrm{coup}}
=
-\frac12
\left[
\sum_j J_{kj}(s'_k s'_j - s_k s_j)
+
\sum_i J_{ik}(s'_i s'_k - s_i s_k)
\right].
$$

Ponieważ dla $j\neq k$ mamy $s'_j=s_j$, a $s'_k=-s_k$, dostajemy

$$
s'_k s'_j - s_k s_j
=
(-s_k)s_j - s_k s_j
=
-2s_k s_j.
$$

Analogicznie,

$$
s'_i s'_k - s_i s_k
=
s_i(-s_k)-s_i s_k
=
-2s_i s_k.
$$

Zatem

$$
\Delta E_{\mathrm{coup}}
=
-\frac12
\left[
\sum_j J_{kj}(-2s_k s_j)
+
\sum_i J_{ik}(-2s_i s_k)
\right].
$$

Wyciągamy wspólne czynniki:

$$
\Delta E_{\mathrm{coup}}
=
\frac12
\left[
2s_k\sum_j J_{kj}s_j
+
2s_k\sum_i J_{ik}s_i
\right].
$$

Czyli

$$
\Delta E_{\mathrm{coup}}
=
s_k\sum_j J_{kj}s_j
+
s_k\sum_i J_{ik}s_i.
$$

Macierz sprzężeń jest symetryczna, czyli

$$
J_{ij}=J_{ji},
$$

więc oba człony są równe i dostajemy

$$
\Delta E_{\mathrm{coup}}
=
2s_k\sum_j J_{kj}s_j.
$$

## 4. Pola zewnętrzne (h)

Dla pola zewnętrznego zostaje tylko składnik z $i=k$:

$$
\Delta E_{\mathrm{field}}
=
-\;h_k(s'_k-s_k).
$$

Ponieważ $s'_k=-s_k$, mamy

$$
s'_k-s_k=-s_k-s_k=-2s_k,
$$

więc

$$
\Delta E_{\mathrm{field}}
=
-\,h_k(-2s_k)
=
2h_k s_k.
$$

## 5. Całkowita różnica energii

Dodając obie części:

$$
\Delta E
=
\Delta E_{\mathrm{coup}}+\Delta E_{\mathrm{field}}
=
2s_k\sum_j J_{kj}s_j + 2h_k s_k.
$$

Wyłączamy $2s_k$ przed nawias:

$$
\boxed{
\Delta E = E(s')-E(s)
=
2s_k\left(h_k + \sum_j J_{kj}s_j\right)
}
$$
