# Wyprowadzenie wzorów:

## Równanie ruchu

Wyobraź sobie masę na sprężynie: gdy ją odchylimy od położenia równowagi, sprężyna „ciągnie” ją z powrotem. W idealnym (liniowym) przybliżeniu siła sprężyny $F_s$ jest proporcjonalna do wychylenia:

$$
F_s = -kx(t),
$$

gdzie znak $-$ oznacza, że siła działa przeciwnie do wychylenia (zawsze w stronę równowagi). Liczbę $k$ nazywamy stałą sprężystości i określa jak "silna" jest sprężyna.

Z drugiej zasady Newtona mamy, że siła to masa $\times$ przyspieszenie. Przyspierzenie $a$ to druga pochodna po położeniu $x(t)$ więc:

$$
F_n = ma \, \Rightarrow \, F_n = m \ddot{x}(t) 
$$

W tym uproszczonym układzie jedyną siłą która działa jest siła sprężyny. Podstawiając mamy:

$$
F_n = F_s = -kx(t)
$$

$$
m \ddot{x}(t) = -kx(t)
$$

Przekształcając równanie, otrzymujemy:


$$
\ddot{x}(t) + \frac{k}{m}x(t) = 0
$$

Często się wprowadza parametr $\omega$:

$$
\omega = \sqrt{\frac{k}{m}},
$$

który nazywamy częstością kołową (lub kątową). Dostajemy standardową postać:

$$
\ddot{x}(t) + \omega^2x(t) = 0
$$




## Energia


Całkowita energia $E$ to suma energii kinetycznej $E_k$ i energi potencjalnej $E_p$.   
$$
E = E_k + E_p
$$

W naszym modelu

$$
E_k = \frac{1}{2}mv^2 = \frac{1}{2} m \dot{x}^2
$$

$$
E_p = \frac{1}{2} \omega^2 x^2
$$

Przyjmując masę jednostkową $m=1$ otrzymujemy:

$$
E = \frac{1}{2}\dot{x}^2 + \frac{1}{2}\omega^2 x^2
$$
