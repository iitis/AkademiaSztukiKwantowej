## 2. Weryfikacja: czy `julia` jest dostępna w PATH


```python
import shutil, subprocess
print("Ścieżka do 'julia':", shutil.which('julia'))
try:
    out = subprocess.check_output(['julia','--version'], text=True)
    print(out.strip())
except Exception as e:
    print("Nie udało się uruchomić 'julia'. Upewnij się, że zakończyłeś instalację i że 'julia' jest w PATH.")
```

## 3. Instalacja kernela IJulia dla Jupyter

IJulia rejestruje kernel automatycznie po instalacji. Wykonaj:


```python
import subprocess
cmd = ['julia','-e','using Pkg; Pkg.add("IJulia"); using IJulia']
print('>',' '.join(cmd))
subprocess.run(cmd, check=True)
```

## 4. Wybór paczek do instalacji

Edytuj listę `packages` i uruchom komórkę. Paczki zostaną zainstalowane globalnie (w domyślnym środowisku Julia).
Jeżeli chcesz pracować w osobnym projekcie, zobacz sekcję 5B.


```python
# Edytuj tę listę według potrzeb:
packages = [
    "DataFrames",
    "CSV",
    "Plots",
]

import json, shlex, subprocess
julia_expr = f'using Pkg; Pkg.add({json.dumps(packages)})'
cmd = ['julia','-e', julia_expr]
print("> julia -e", shlex.quote(julia_expr))
subprocess.run(cmd, check=True)
```

## 5A. Weryfikacja zainstalowanych paczek w środowisku domyślnym


```python
import subprocess
subprocess.run(['julia','-e','using Pkg; Pkg.status()'], check=True)
```

## 5B. (Opcjonalnie) Projekt Julia z własnym środowiskiem

Poniższe komendy tworzą katalog projektu, aktywują środowisko i instalują paczki tylko dla tego projektu.


```python
project_name = "MojeSrodowiskoJulia"
packages_project = [
    "DataFrames",
    "CSV",
]

import json, subprocess, shlex
expr = (
    f'using Pkg; '
    f'Pkg.activate("{project_name}", shared=false); '
    f'Pkg.add({json.dumps(packages_project)}); '
    f'Pkg.precompile(); '
    f'Pkg.status()'
)
print("> julia -e", shlex.quote(expr))
subprocess.run(['julia','-e', expr], check=True)
```

## 6. Test działania

Krótki test: wypisanie wersji Julia oraz użycie paczek.


```python
import subprocess, textwrap
test_expr = textwrap.dedent('''
    using InteractiveUtils
    versioninfo()
    using DataFrames, CSV, Plots
    println("OK: załadowano DataFrames, CSV, Plots")
''').strip()
subprocess.run(['julia','-e', test_expr], check=True)
```
