# Julia

Julia jest jednym z popularniejszych języków programowania 

instrukcje jak zainstalować julie można znaleźć [tutaj](https://julialang.org/install/)

Nie będziemy tworzyć wirtualnego środowiska

uruchom Julia w terminalu/wierszu poleceń poprzez wpisanie komendy:

Julia

Następnie aktywujemy managera pakietów poprzez naciśnięcie `]` i wpisujemy

add IJulia

add CUDA

add TensorOperations

Teraz notebook powinien działać. Podobnie jak w przypadku pythona dodajemy wybieramy by kernelem była Julia
